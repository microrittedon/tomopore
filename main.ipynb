```python
import tomopore.pordiamdist as tp
```

```python
im = tp.create_phantom(blobiness=6,porosity=0.1)
```

```python
tp.reconstruction_parameters(number_of_angles=50,threshold = 0.4,rec_algorithm='FBP')
```

```python
sin = tp.create_sinogram(im)
```

```python
imr=tp.rescale(tp.reconstruct(sin))
```

```python
tp.show.show_distributions_of_diameters(im,imr,WriteMarkdown=True)
```

```python
def process():
    tp.reconstruction_parameters(number_of_angels=50,threshold = 0.4,rec_algorithm='FBP')
    sin = tp.create_sinogram(im)
    imr=tp.rescale(tp.reconstruct(sin))
    tp.show.show_distributions_of_diameters(im,imr,WriteMarkdown=True)
```
